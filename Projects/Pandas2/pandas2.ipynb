```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Problem 1

```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """
    df = pd.read_csv('crime_data.csv')
    mean_df = df.mean()
    rate_df = df['Total'] / df['Population']

    # Create a figure and a 3x1 grid of subplots
    fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 10))

    # Plotting the first graph on the first subplot
    df.plot(x='Year', y=['Violent', 'Property'], linewidth=1, title="Property Crime vs Violent Crime", ax=axes[0])

    # Plotting the second graph on the second subplot
    mean_df.drop(['Year', 'Population', 'Total']).plot(kind="barh", title="Mean of Various Crimes", ax=axes[1])

    rate_df.plot(kind="hist",bins=10, grid=False, title="Distribution of Crime Rates", ax=axes[2])
    
    # Show the plots
    plt.tight_layout()
    plt.show()
prob1()
```

# Problem 2

```python
def prob2():
    """
    Use crime_data.csv to plot the trends between Larceny and
        1. Violent
        2. Burglary
        3. Aggravated Assault
    Each graph should be clearly labelled and readable.
    One of these variables does not have a linear trend with Larceny.
    Return a string identifying this variable.
    """
    # read in data
    df = pd.read_csv("crime_data.csv")
    
    # plot the correlations in a scatter plot
    df.plot(kind='scatter', y='Larceny', x='Violent')
    df.plot(kind='scatter', y='Larceny', x='Burglary')
    df.plot(kind='scatter', y='Larceny', x='Aggravated Assault')


    plt.tight_layout()
    plt.show()

    # return the crime with the lowest correlation
    return df[['Larceny','Violent','Burglary','Aggravated Assault']].corr()['Larceny'].idxmin()


prob2()
```

# Problem 3

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft 
            as box plots
        2. The distributions of Vehicle Thefts against Robberies as
            a hexbin plot.
    All plots should be labelled and easy to read.
    """
    # read in data
    df = pd.read_csv('crime_data.csv')

    # create box plot
    df.plot(kind='box', y=['Burglary','Violent','Vehicle Theft'], title="Box Plot of Burglary, Violent, and Vehicle Theft")

    # create hexbin plot
    df.plot(kind='hexbin', x="Vehicle Theft", y="Robbery", gridsize=15, sharex=False, title="Hexbin Plot of Vehicle Thefts against Robbery")

    # plot graph
    plt.tight_layout()
    plt.show()

prob3()
```
